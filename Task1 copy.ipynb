```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python

x=[]
y=[]

x2=[]
y2=[]
# zmienne
x0, p0 = 3, 1.8999
x02,p02 = 3, 1.9
K=1.2

x.append(x0)
y.append(p0)

x2.append(x02)
y2.append(p02)

n=np.linspace(0,1,50)
# calculate x,p points depending on initial values
def funkcja(xn,pn):
    p1=pn+K*math.sin(xn)
    x1=(xn+p1)%(2*math.pi)
    return x1,p1

for i in range(len(n)):
    a,b=funkcja(x[i],y[i])
    c,d=funkcja(x2[i],y2[i])
    x.append(a)
    y.append(b)
    x2.append(c)
    y2.append(d)



```

```python

plt.title("Dimensionless equation of motion initial points p0=1.8999, x0=3, K=1.2", fontsize=12)

plt.subplot(2,1,1)
plt.plot(x,label="Xn values")
plt.plot(y,label="Pn values")
plt.xlabel("number of kicks")
plt.ylabel("returned values")
plt.title("P and X values for dimensionless equation of motion", fontsize=12)
plt.legend(loc="best")
plt.grid(True)





plt.subplot(2,1,2)
plt.plot(x2,label="Xn values")
plt.plot(y2,label="Pn values")
plt.xlabel("number of kicks")
plt.ylabel("returned values")
plt.title("Dimensionless equation of motion initial points p0=1.3, x0=3, K=1.2", fontsize=12)
plt.grid(True)
plt.legend(loc="best")

plt.savefig("who_lstrajectories.png")
```
