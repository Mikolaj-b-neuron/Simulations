```python

# Define the size of the big array
size = 50

# Generate a random big array with some ones put inside
big_array = np.zeros((size, size))
big_array[10:13, 10:13] = 1
big_array[20:23, 20:23] = 1

# i put it straight for now
window_size = 3
# we create new blank one in which we will be creating
output_array = np.zeros((size, size))

for i in range(size - window_size + 1):
    for j in range(size - window_size + 1):
        # we cut out small window i which we look for match with filter
        window = big_array[i:i+window_size, j:j+window_size]

        # Now that we cut out our window, we check if it covers our 'filter' named arr
        if np.array_equal(window, arr) == True:
            # If they match, set the center of the window in the output array to a new value
            output_array[i+1, j+1] = 1

plt.figure()
plt.imshow(big_array, cmap='binary')
plt.grid()
plt.show()

plt.figure()
plt.imshow(output_array, cmap='binary')
plt.grid()
plt.show()
```

```python
# Define the size of the grid
n = 10
m = 10

# Create a 2D array to represent the grid of cells
grid = np.zeros((n, m))

# Define the initial state of each cell in the array
grid[n//2, m//2] = 1
grid[n//2,m//2+1] = 1
grid[n//2,m//2-1] = 1
grid[n//2 - 1,m//2] = 1


def update(grid):
    new_grid = np.zeros_like(grid)
    for i in range(n):
        for j in range(m):

            count = np.sum(grid[max(0,i-1):min(n,i+2), max(0,j-1):min(m,j+2)]) - grid[i,j]
            if grid[i,j] == 1 and (count == 2 or count == 3):
                new_grid[i,j] = 1
            elif grid[i,j] == 0 and count == 3:
                new_grid[i,j] = 1
    return new_grid

# Run the simulation for a fixed number of time steps
for i in range(0,5):
    grid = update(grid)
    plt.figure()
    plt.imshow(grid, cmap='binary')
    plt.grid()
    plt.show()

#plt.figure()
#plt.imshow(grid, cmap='binary')
#plt.grid()
#plt.show()

     
```
