```python
# Imported libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy 
```

```python
M=100
n=np.arange(M)
h=2*np.pi/M
xn=np.zeros(M)
x=np.zeros(M)
for i in n:
    xn[i]=(2*np.pi*n[i]/M)
    

hbar=2*np.pi/M
#print(n)
```

```python
def continuumExpression(x,x0,p0):
    
    psi=np.exp(1j* (p0*x)/(h))*np.exp((-(x-x0)**2)/(2*h))
    psi=psi/np.linalg.norm(np.exp(1j* (p0*x)/(h))*np.exp((-(x-x0)**2)/(2*h)))

    return psi

def GaussianWave(x,x0,p0):
    test=0
    for i in range(-4,4):
        test+= np.exp(-((x-x0+2*np.pi*i)**2)/(2*h))
    
    psi1=np.exp(1j* (p0*x)/(h))*test
    psi1=psi1/np.linalg.norm(psi1)
    return psi1

#print(GaussianWave(xn,10,10))
```

```python
# Part 1 - Gaussian wave for x0=20
plt.figure()
plt.plot(GaussianWave(xn,xn[20],xn[40]))
plt.axvline(x=20, color='g', linestyle=':', linewidth=3)
plt.xlabel("x value")
plt.show()
```

```python
# Part 2 = |psi|**2
psi_abs=GaussianWave(xn,xn[10],xn[40])
psi_abs=np.abs(psi_abs)**2
plt.figure()
plt.plot(psi_abs)
plt.xlabel('x value')
plt.axvline(x=10, color='g', linestyle=':', linewidth=3)
plt.show()
```

```python
''' Part 3 - Perform the fast Fourier transform (FFT) on the vector G(xn).
Plot the resulting wave function amplitude, check normalisation,
mark the value p0 '''
psi_fft=scipy.fft.fft(GaussianWave(xn,xn[10],xn[40]))
#p=2*np.pi*n/M
plt.figure()
plt.plot(np.abs(psi_fft)**2)
plt.xlabel('p value')
plt.axvline(x=40, color='r', linestyle=':', linewidth=3)
```
