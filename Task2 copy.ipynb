```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python
# calculate x,p points depending on initial values
def funkcja(xn,pn,K):
    p1=(pn+K*math.sin(xn))%(2*math.pi)
    x1=(xn+p1)%(2*math.pi)
    return x1,p1
```

```python
K1=1.2
K2=2.1
K3=5.5
N=np.linspace(0,1000,1)
```

```python
# otweiramy macierze do zapisu x i p
x=[]
p=[]
# Initial values for x and N are random
x0, p0 = np.random.random(2)*2*np.pi
x.append(x0)
p.append(p0)

# 1000 kicks
for i in range(1000):
    a,b=funkcja(x[i],p[i],K1)
    x.append(a)
    p.append(b)
# plot results
plt.plot(x,p)
#plt.plot(p)
```

```python
def trajeciories(K):
    
    for i in range(100):
        #r, g, b = np.random.random(3) 
        x=[]
        p=[]
        x0, p0 = np.random.random(2)*2*np.pi
        x.append(x0)
        p.append(p0)

        # 1000 kicks
        for i in range(1000):
            a,b=funkcja(x[i],p[i],K)
            x.append(a)
            p.append(b)
    # plot results
        plt.plot(x,p,',')

plt.figure(figsize=(5, 3), dpi=150)
plt.subplot(3,1,1)
plt.title("trajectorie for K= 1.2")
trajeciories(K=K1)
plt.subplot(3,1,2)
plt.title("trajceotrie for K=2.1")
trajeciories(K=K2)
plt.subplot(3,1,3)
plt.title("trajceotrie for K=5.5")
trajeciories(K=K3)
```
